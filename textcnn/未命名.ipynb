```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt 

from scipy.interpolate import spline

xxx=[1,2,3,4,5,6,7,8]
# xxx=['4W','6W','8W','10W','12W']

T = np.array([1,2,3,4,5,6,7,8])

xnew = np.linspace(T.min(),T.max(),300)
xy=[0.6875,0.6875,0.7188,0.7500,0.7500,0.7812,0.75,0.68]

p1=spline(T,xy,xnew)

plt.figure(figsize=(8,6))
plt.plot(xnew,p1, 'b',label='acc')
P2=plt.scatter(xxx,xy,c='r',marker='*',lw=2)
plt.legend(loc='best',fontsize=17)
plt.grid(b=True, ls=':')
plt.show()
```
